```python
# import all the basic packges for the generation of WordCloud
from PIL import Image
import wikipedia
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
print ('Wordcloud is installed and imported!')
```

```python
# open the file and read it into a variable alice_novel
alice_novel = wikipedia.page("Alice's Adventures in Wonderland")
    
print ('File downloaded and saved!')
```

```python
stopwords = set(STOPWORDS)
len(stopwords), len(alice_novel.content)
```

```python
# instantiate a word cloud object
alice_wc = WordCloud(
    background_color='white',
    stopwords=stopwords
)

# generate the word cloud
alice_wc.generate(alice_novel.content)
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
alice_mask = np.array(Image.open('Mask.png'))
    
print('Image downloaded and saved!')
```

```python
# Plot the mask
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
alice_wc = WordCloud(background_color='white', mask=alice_mask, stopwords=stopwords)

# generate the word cloud
alice_wc.generate(alice_novel.content)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
